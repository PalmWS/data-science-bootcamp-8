```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# IMDb URL of the top-rated movies page
url = "https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc"

# Send an HTTP GET request to the IMDb page
response = requests.get(url)

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Find all movie containers
movie_containers = soup.find_all("div", class_="lister-item-content")

# Create lists to store movie data
titles = []
ratings = []

# Extract movie titles and ratings
for container in movie_containers:
    # Find movie title
    title = container.find("h3", class_="lister-item-header").find("a").text.strip()
    titles.append(title)

    # Find movie rating
    rating = container.find("div", class_="inline-block ratings-imdb-rating").find("strong").text.strip()
    ratings.append(rating)

# Create a DataFrame
imdb_data = pd.DataFrame({"Title": titles, "Rating": ratings})

# Display the DataFrame
imdb_data
```
