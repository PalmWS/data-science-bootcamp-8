```python
import requests
import time
import pandas as pd

# Define the API URL
base_url = "https://pokeapi.co/api/v2/pokemon/"

# Number of Pokémon you want to retrieve (e.g., 10)
num_pokemon = 10

# Initialize an empty list to store Pokémon data
pokemon_data = []

# Fetch data for each Pokémon
for pokemon_id in range(1, num_pokemon + 1):
    response = requests.get(f"{base_url}{pokemon_id}")
    time.sleep(1)

    if response.status_code == 200:
        data = response.json()

        # Extract relevant information
        pokemon_info = {
            "Name": data["name"],
            "Height": data["height"],
            "Weight": data["weight"],
            "Abilities": [ability["ability"]["name"] for ability in data["abilities"]],
            "Types": [type_entry["type"]["name"] for type_entry in data["types"]]
        }

        # Append Pokémon info to the list
        pokemon_data.append(pokemon_info)


# Create a Pandas DataFrame from the list of Pokémon data
df = pd.DataFrame(pokemon_data)

# Display the DataFrame
df
```
